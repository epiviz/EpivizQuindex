## Usage

create index

```python
from epivizquindex import EpivizQuindex
from epivizquindex.utils import get_genome
import pandas as pd
import seaborn as sns
pd.set_option('display.width', 1000)
```

```python
genome = get_genome('mm10')

index = EpivizQuindex.EpivizQuindex(genome, base_path='./showcase_index/')
f1 = 'http://renlab.sdsc.edu/yangli/downloads/mousebrain/bigwig/VPIA3.bw'
f2 = 'http://renlab.sdsc.edu/yangli/downloads/mousebrain/bigwig/PVM.bw'
f3 = 'http://renlab.sdsc.edu/yangli/downloads/mousebrain/bigwig/OBGL4.bw'
f4 = 'http://renlab.sdsc.edu/yangli/downloads/mousebrain/bigwig/OBGL5.bw'

# adding file to index
index.add_to_index(f1)
index.add_to_index(f2)
index.add_to_index(f3)
index.add_to_index(f4)

```

Query

```python
# querying a range in 1 file
index.query("chrX", 0, 3195790, file_name = f1)
```

```python
# querying for a range in all files
index.query("chrX", 0, 3195790)
```

Save index to disk

```python
index.to_disk()
```

Load from disk

```python
index = EpivizQuindex.EpivizQuindex(genome, base_path='./showcase_index/')
index.from_disk()
```

```python
index.query("chrX", 0, 3195790, file_name = f1)
```

Perform on-disk query

```python
index = EpivizQuindex.EpivizQuindex(genome, base_path='./showcase_index/')
index.from_disk(load = False)

index.query("chrX", 0, 3195790, file_name = f1, in_memory = False)
```

```python
index.has_data("chrX", 0, 3195790, in_memory=False)
```

## Plot heatmap

```python
start = 3193100
end = 3195790
records = index.query("chrX", start, end, in_memory = False)
records
```

```python
entries = {}
for file_name in records.file_name.unique():
    entries[file_name] = []
    for x in range(start, end + 1):
        tb = records.loc[records['start'] <= x]
        tb = tb.loc[tb['end'] > x]
        value = list(tb.loc[tb['file_name'] == file_name]['score'])[0]
        entries[file_name].append(value)
    
values = pd.DataFrame(entries).transpose()
values.columns = [x for x in range(start, end+1)]
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.heatmap(values)
```
